# Ejercicio 1

```python
def promedio(combustibleTanques):
    totalTanques = sum(combustibleTanques)
    numTanques = len(combustibleTanques)
    return totalTanques / numTanques
def informe(tanque1, tanque2, tanque3):
 
    return f"""Informe de combustible:
    Promedio: {promedio([tanque1, tanque2, tanque3])}%
    Tanque 1: {tanque1}%
    Tanque 2: {tanque2}%
    Tanque 3: {tanque3}% 
    """

print(informe(88, 76, 70))

promedio([80, 85, 81]) 
```

# Ejercicio 2

```python
def informeCohete(destino, *minutes, **fuel_reservoirs):
    informe = f"""
    Destino: {destino}
    Tiempo de vuelo: {sum(minutes)} minutos
    Combustible restante: {sum(fuel_reservoirs.values())} litros
    """
    for tanqueNombre, litros in fuel_reservoirs.items():
        informe += f"{tanqueNombre} Tanque --> {litros} Litros restantes\n"
    return informe

print(informeCohete("Moon", 10, 15, 51, main=300000, external=200000))

```

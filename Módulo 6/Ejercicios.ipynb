# Ejercicio 2

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
planetaBuscar = input('Ingrese el nombre del planeta (La primera letra en mayúscula, por favor):')

x = planets.index(planetaBuscar)

print('Estos son los planetas más cercanos a: ' + planetaBuscar)
print(planets[0:x])

print('Estos son los planetas más lejanos a: ' + planetaBuscar)
print(planets[x + 1:])
```
